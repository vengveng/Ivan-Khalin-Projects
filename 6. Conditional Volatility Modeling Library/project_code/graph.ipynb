```python
import numpy as np
from scipy.stats import jf_skew_t, norm
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

# Parameters for jf_skew_t distribution
a, b = 3.016092, 3.014395

# Use a wide linspace for both distributions
x = np.linspace(-6, 6, 1000)

# Plot the jf_skew_t distribution over the wide range
ax.plot(x, jf_skew_t.pdf(x, a, b),
       'r-', lw=5, alpha=0.6, label='jf_skew_t pdf')

# Frozen jf_skew_t pdf
rv = jf_skew_t(a, b)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Plot the standard normal distribution
ax.plot(x, norm.pdf(x),
        'b--', lw=2, label='normal (0, 1) pdf')

# Add labels and legend
ax.legend(loc='best')
ax.set_title('Comparison of jf_skew_t and Normal Distributions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filepaths for the data files
filepaths = [
    '/Users/ivankhalin/Documents/code/MA3/QARM/garchlib/non_diagonal_correlations_garch_studentt.csv',
    '/Users/ivankhalin/Documents/code/MA3/QARM/garchlib/non_diagonal_correlations_studentt.csv',
    '/Users/ivankhalin/Documents/code/MA3/QARM/garchlib/non_diagonal_correlations_normal_garch.csv',
    '/Users/ivankhalin/Documents/code/MA3/QARM/garchlib/non_diagonal_correlations_normal.csv'
]

# Create a large plot with Matplotlib
plt.figure(figsize=(15, 8))  # Set a large size for the plot

# Iterate through each file in filepaths and plot the data
for filepath in filepaths:
    # Load the data
    df = pd.read_csv(filepath)
    
    # Plot the non-diagonal correlations
    plt.plot(df.index, df['Non_Diagonal_Correlation'], label=filepath.split('/')[-1], linewidth=2)

# Add title and labels
plt.title('Non-Diagonal Correlations Over Time from Different Models', fontsize=20)
plt.xlabel('Time Steps', fontsize=15)
plt.ylabel('Non-Diagonal Correlation', fontsize=15)

# Show the minimum and maximum correlation values from the last dataset for debugging purposes
print(min(df['Non_Diagonal_Correlation']), max(df['Non_Diagonal_Correlation']))

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
